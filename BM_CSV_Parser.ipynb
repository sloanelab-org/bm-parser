#### Declare prefixes and create graph

```python
import json
with open(f'{DATA_PATH}bm_img.json') as f:
    img = json.load(f)
    
IMG_URL = 'https://media.britishmuseum.org/media/'

with open(f'{DATA_PATH}sloane_ppl.json') as g:
    ppl = json.load(g)

known_people = {}
    
for p in ppl:
    print(p.split('/E21/')[1])
    known_people[p.split('/E21/')[1]] = p
```

#### Add catalogue

```python
import rfc3987
import Levenshtein

# Create namespaces
slb=Namespace("http://sloanelab.org/");
ecrm=Namespace("http://erlangen-crm.org/current/");
crmdig=Namespace("http://www.ics.forth.gr/isl/CRMdig/");

BATCH = 3000
j = 0

while j < 16000:    
    # Create graph
    graph = rdflib.Graph()

    # Bind prefixes
    graph.bind("slb", slb)
    graph.bind("ecrm", ecrm)
    graph.bind("crmdig", crmdig)

    # Dataset
    datasetIRI = slb[f'D1/bm_dataset'];
    graph.add((datasetIRI, RDF.type, crmdig.D1_Digital_Object))
    graph.add((datasetIRI, ecrm.P2_has_type, slb['E55/museum_dataset']))
    graph.add((datasetIRI, RDFS.label, Literal('BM Dataset')))
    
    museumIRI = slb[f'E74/British_Museum']
    graph.add((museumIRI, RDFS.label, Literal(f'British Museum')))

    for i, loc in enumerate(range(j,min(j+BATCH, 15402))):
        
        if (i > 15402):
            break
        
        row = bm_dataset.loc[loc]

        # Record
        recordID = row["Unique ID"]
        recordIRI = slb[f'E73/bm_dataset/{recordID}'];
        graph.add((recordIRI, RDF.type, ecrm.E73_Information_Object))
        graph.add((recordIRI, ecrm.P2_has_type, slb['E55/museum_record']))    
        graph.add((recordIRI, RDFS.label, Literal(f'BM Record {recordID}')))
        graph.add((recordIRI, ecrm.P48_has_preferred_identifier, Literal(recordID)))
        graph.add((datasetIRI, ecrm.P106_is_composed_of, recordIRI))

        objID = row["Unique ID"]
        objIRI = slb[f'E19/bm_dataset/{objID}'];
        #obj_label = bm_dataset.loc[loc]["Object Name - Object Name"]
        obj_desc = bm_dataset.loc[loc]["Physical description"]
        #obj_id = bm_dataset.loc[loc]["Registration number"]
        
        # Link to Historical Catalogue
        histCat = row["Registration Number - Collection"]
        histNum = row["Registration Number - Number"]
        
        if not pd.isna(histCat) and not pd.isna(histNum):
            histIRI = ""
            if (histCat.strip().split('~')[0] == 'SLMisc'):
                histIRI = slb[f'E19/miscellanies/{histNum.strip()}']
            elif (histCat.strip().split('~')[0] == 'SLSeals'):
                histIRI = slb[f'E19/seals/{histNum.strip()}']
            elif (histCat.strip().split('~')[0] == 'SLPictures'):
                histIRI = slb[f'E19/pictures/{histNum.strip()}']
            elif (histCat.strip().split('~')[0] == 'SLHandles'):
                histIRI = slb[f'E19/agate_handles/{histNum.strip()}']
            elif (histCat.strip().split('~')[0] == 'SLRings'):
                histIRI = slb[f'E19/rings/{histNum.strip()}']
            elif (histCat.strip().split('~')[0] == 'SLCups'):
                histIRI = slb[f'E19/agate_cups_bottles_spoons/{histNum.strip()}']
            elif (histCat.strip().split('~')[0] == 'SLMathInstr'):
                histIRI = slb[f'E19/mathematical_instruments/{histNum.strip()}']
            elif (histCat.strip().split('~')[0] == 'SLAntiq'):
                histIRI = slb[f'E19/antiquities/{histNum.strip()}']
            
            if histIRI:
                graph.add((recordIRI, ecrm.P67_refers_to, Literal(histIRI)))
                graph.add((objIRI, slb.believed_to_be_the_same_as, histIRI))
        
        if objID in img and img[objID]:
            graph.add((URIRef(f'{IMG_URL}{img[objID]}'), ecrm.P138_represents, objIRI))

        #graph.add((recordIRI, RDFS.label, Literal(obj_label)))
        graph.add((recordIRI, RDFS.comment, Literal(obj_desc)))
        graph.add((recordIRI, ecrm.P190_has_symbolic_content, Literal(obj_desc)))
        
        if not pd.isna(row["Comment"]):
            graph.add((recordIRI, slb.has_curator_comment, Literal(row["Comment"].strip())))

        graph.add((objIRI, RDFS.label, Literal(f'BM Object {objID}')))
        graph.add((objIRI, RDF.type, ecrm.E19_Physical_Object))
        graph.add((objIRI, ecrm.P48_has_preferred_identifier, Literal(objID)))
        graph.add((recordIRI, ecrm.P129_is_about, objIRI))
        
        graph.add((objIRI, RDFS.comment, Literal(f'An object from the BM dataset')))
        graph.add((objIRI, ecrm['P50_has_current_keeper'], museumIRI))

        catItemCount = 1

        # For each person...
        for person in str(row["Producer/Artist - Name"]).split('~'):
            
            if person.lower() == 'nan' or person.lower() == 'anonymous':
                continue
            
            if ', ' in person:
                name_parts = person.split(', ')
                person = f'{name_parts[1]}_{name_parts[0]}'
            
            matches = []
            
            for p in known_people.keys():
                r = Levenshtein.ratio(p, person)
                if r > .9:
                    #print(p)
                    #print(person)
                    #print(r)
                    matches.append((r, known_people[p]))
            
            if matches:
                personIRI = URIRef(sorted(matches, reverse=True)[0][1])
                #print(personIRI)
            else:
                # Create person IRI
                personIRI = slb[f'E21/{person.strip().replace(" ", "_")}'];
                if not rfc3987.parse(personIRI)['IRI']:
                    print(personIRI)
                graph.add((personIRI, RDFS.label, Literal(person.strip().replace("_", " "))))
            graph.add((personIRI, RDF.type, ecrm.E21_Person))

            # Create catalogue item IRI
            catItemIRI = slb[f'E73/bm_dataset/{recordID}/{catItemCount}'];
            catItemCount += 1
            graph.add((catItemIRI, RDFS.label, Literal(f'BM catalogue item {recordID}/{catItemCount}')))

            # Connect catalogue item to record
            graph.add((recordIRI, ecrm.P106_is_composed_of, catItemIRI))

            # Connect person to record
            graph.add((catItemIRI, slb.SL10_specifically_refers_to, personIRI))

            assoc = str(row["Producer/Artist - Association"])

            if assoc:
                assocIRI = slb[f'E21/{assoc.strip().replace(" ", "_")}'];
                if not rfc3987.parse(assocIRI)['IRI']:
                    print(assocIRI)
                graph.add((catItemIRI, slb.SL11_has_specific_type, assocIRI))

        # For each person...
        for person in str(row["Excavator/Field collector - Name"]).split('~'):
            
            if person.lower() == 'nan' or person.lower() == 'anonymous':
                continue
            
            if ', ' in person:
                name_parts = person.split(', ')
                person = f'{name_parts[1]}_{name_parts[0]}'

            matches = []
            
            for p in known_people.keys():
                r = Levenshtein.ratio(p, person)
                if r > .9:
                    #print(p)
                    #print(person)
                    #print(r)
                    matches.append((r, known_people[p]))
            
            if matches:
                personIRI = URIRef(sorted(matches, reverse=True)[0][1])
                #print(personIRI)
            else:
                # Create person IRI
                personIRI = slb[f'E21/{person.strip().replace(" ", "_")}'];
                if not rfc3987.parse(personIRI)['IRI']:
                    print(personIRI)
                graph.add((personIRI, RDFS.label, Literal(person.strip().replace("_", " "))))
            graph.add((personIRI, RDF.type, ecrm.E21_Person))

            # Create catalogue item IRI
            catItemIRI = slb[f'E73/bm_dataset/{recordID}/{catItemCount}'];
            catItemCount += 1
            graph.add((catItemIRI, RDFS.label, Literal(f'BM catalogue item {recordID}/{catItemCount}')))

            # Connect catalogue item to record
            graph.add((recordIRI, ecrm.P106_is_composed_of, catItemIRI))

            # Connect person to entry
            graph.add((catItemIRI, slb.SL10_specifically_refers_to, personIRI))

            assoc = str(row["Excavator/Field collector - Association"])

            if assoc:
                assocIRI = slb[f'E21/{assoc.strip().replace(" ", "_")}'];
                if not rfc3987.parse(assocIRI)['IRI']:
                    print(assocIRI)
                graph.add((catItemIRI, slb.SL11_has_specific_type, assocIRI))

        # For each place...
        for place in str(row["Production place"]).split('~'):
            
            if pd.isna(row["Production place"]):
                continue

            if place.strip().replace(" ", "_"):
                # Create place IRI
                try:
                    placeIRI = slb[f'E53/{place.strip().replace(" ", "_")}'];
                    if not rfc3987.parse(placeIRI)['IRI']:
                        print(placeIRI)
                    graph.add((placeIRI, RDF.type, ecrm.E53_Place))
                    graph.add((placeIRI, RDFS.label, Literal(place.strip().replace("_", " "))))
                except AttributeError:
                    print(place["id"])

                # Create catalogue item IRI
                catItemIRI = slb[f'E73/bm_dataset/{recordID}/{catItemCount}'];
                catItemCount += 1
                graph.add((catItemIRI, RDFS.label, Literal(f'BM catalogue item {recordID}/{catItemCount}')))

                # Connect catalogue item to record
                graph.add((recordIRI, ecrm.P106_is_composed_of, catItemIRI))

                # Connect place to entry
                graph.add((catItemIRI, slb.SL10_specifically_refers_to, placeIRI))

                assoc = str(row["Production place - Association"])

                if assoc:
                    assocIRI = slb[f'E21/{assoc.strip().replace(" ", "_")}'];
                    if not rfc3987.parse(assocIRI)['IRI']:
                        print(assocIRI)
                    graph.add((catItemIRI, slb.SL11_has_specific_type, assocIRI))

        # For each place...
        for place in str(row["Findspot - Place"]).split('~'):
            
            if pd.isna(row["Findspot - Place"]):
                continue

            if place.strip().replace(" ", "_"):
                # Create place IRI
                try:
                    placeIRI = slb[f'E53/{place.strip().replace(" ", "_")}'];
                    if not rfc3987.parse(placeIRI)['IRI']:
                        print(placeIRI)
                    graph.add((placeIRI, RDF.type, ecrm.E53_Place))
                    graph.add((placeIRI, RDFS.label, Literal(place.strip().replace("_", " "))))
                except AttributeError:
                    print(place["id"])

                # Create catalogue item IRI
                catItemIRI = slb[f'E73/bm_dataset/{recordID}/{catItemCount}'];
                catItemCount += 1
                graph.add((catItemIRI, RDFS.label, Literal(f'BM catalogue item {recordID}/{catItemCount}')))

                # Connect catalogue item to record
                graph.add((recordIRI, ecrm.P106_is_composed_of, catItemIRI))

                # Connect place to entry
                graph.add((catItemIRI, slb.SL10_specifically_refers_to, placeIRI))

                assoc = str(row["Findspot - Association"])

                if assoc:
                    assocIRI = slb[f'E21/{assoc.strip().replace(" ", "_")}'];
                    if not rfc3987.parse(assocIRI)['IRI']:
                        print(assocIRI)
                    graph.add((catItemIRI, slb.SL11_has_specific_type, assocIRI))

        # For each material...
        for material in str(row["Material"]).split('~'):
            
            if pd.isna(row["Material"]):
                continue

            if place.strip().replace(" ", "_"):
                # Create material IRI
                try:
                    materialIRI = slb[f'E57/{material.strip().replace(" ", "_")}'];
                    if not rfc3987.parse(materialIRI)['IRI']:
                        print(materialIRI)
                    graph.add((materialIRI, RDF.type, ecrm.E55_Type))
                    graph.add((materialIRI, ecrm.P2_has_type, slb['E55/material']))
                    graph.add((materialIRI, RDFS.label, Literal(material.strip().replace("_", " "))))
                except AttributeError:
                    print(material["id"])
        
                # Create catalogue item IRI
                catItemIRI = slb[f'E73/bm_dataset/{recordID}/{catItemCount}'];
                catItemCount += 1
                graph.add((catItemIRI, RDFS.label, Literal(f'BM catalogue item {recordID}/{catItemCount}')))

                # Connect catalogue item to record
                graph.add((catItemIRI, ecrm.P106_is_composed_of, catItemIRI))

                # Connect material to entry
                graph.add((catItemIRI, slb.SL10_specifically_refers_to, materialIRI))

        # For each technique...
        for technique in str(row["Technique"]).split('~'):
            
            if technique == 'nan':
                continue

            if technique.strip().replace(" ", "_"):
                # Create technique IRI
                try:
                    techniqueIRI = slb[f'E29/{technique.strip().replace(" ", "_")}'];
                    if not rfc3987.parse(techniqueIRI)['IRI']:
                        print(techniqueIRI)
                    graph.add((techniqueIRI, RDF.type, ecrm.E55_Type))
                    graph.add((techniqueIRI, ecrm.P2_has_type, slb['E55/technique']))
                    graph.add((techniqueIRI, RDFS.label, Literal(technique.strip().replace("_", " "))))
                except AttributeError:
                    print(technique["id"])

                # Create catalogue item IRI
                catItemIRI = slb[f'E73/bm_dataset/{recordID}/{catItemCount}'];
                catItemCount += 1
                graph.add((catItemIRI, RDFS.label, Literal(f'BM catalogue item {recordID}/{catItemCount}')))

                # Connect catalogue item to record
                graph.add((recordIRI, ecrm.P106_is_composed_of, catItemIRI))

                # Connect technique to entry
                graph.add((catItemIRI, slb.SL10_specifically_refers_to, techniqueIRI))

        # For each subject...
        for subject in str(row["Subject - Subject"]).split('~'):
            
            if subject == 'nan':
                continue

            if subject.strip().replace(" ", "_"):
                # Create subject IRI
                try:
                    subjectIRI = slb[f'E55/{subject.strip().replace(" ", "_")}'];
                    if not rfc3987.parse(subjectIRI)['IRI']:
                        print(subjectIRI)
                    graph.add((subjectIRI, RDF.type, ecrm.E55_Type))
                    graph.add((subjectIRI, ecrm.P2_has_type, slb['E55/subject']))
                    graph.add((subjectIRI, RDFS.label, Literal(subject.strip().replace("_", " "))))
                except AttributeError:
                    print(subject["id"])

                # Create catalogue item IRI
                catItemIRI = slb[f'E73/bm_dataset/{recordID}/{catItemCount}'];
                catItemCount += 1
                graph.add((catItemIRI, RDFS.label, Literal(f'BM catalogue item {recordID}/{catItemCount}')))

                # Connect catalogue item to record
                graph.add((recordIRI, ecrm.P106_is_composed_of, catItemIRI))

                # Connect subject to entry
                graph.add((catItemIRI, slb.SL10_specifically_refers_to, subjectIRI))

        # For each inscription...
        for inscription in str(row["Inscription type"]).split('~'):
            
            if inscription == 'nan':
                continue

            if inscription.strip().replace(" ", "_"):
                # Create inscription IRI
                try:
                    inscriptionIRI = slb[f'E34/{inscription.strip().replace(" ", "_")}'];
                    if not rfc3987.parse(inscriptionIRI)['IRI']:
                        print(inscriptionIRI)
                    graph.add((inscriptionIRI, RDF.type, ecrm.E55_Type))
                    graph.add((inscriptionIRI, ecrm.P2_has_type, slb['E55/inscription_type']))
                    graph.add((inscriptionIRI, RDFS.label, Literal(inscription.strip().replace("_", " "))))
                except AttributeError:
                    print(subject["id"])

                # Create catalogue item IRI
                catItemIRI = slb[f'E73/bm_dataset/{recordID}/{catItemCount}'];
                catItemCount += 1
                graph.add((catItemIRI, RDFS.label, Literal(f'BM catalogue item {recordID}/{catItemCount}')))

                # Connect catalogue item to record
                graph.add((recordIRI, ecrm.P106_is_composed_of, catItemIRI))

                # Connect inscription to entry
                graph.add((catItemIRI, slb.SL10_specifically_refers_to, inscriptionIRI))

        # For each type...
        for objType in str(row["Object Name - Object Name"]).split('~'):
            
            if objType == 'nan':
                continue

            if objType.strip().replace(" ", "_"):
                # Create type IRI
                try:
                    typeIRI = slb[f'E55/{objType.strip().replace(" ", "_")}'];
                    if not rfc3987.parse(typeIRI)['IRI']:
                        print(typeIRI)
                    graph.add((typeIRI, RDF.type, ecrm.E55_Type))
                    graph.add((typeIRI, ecrm.P2_has_type, slb['E55/object_name']))
                    graph.add((typeIRI, RDFS.label, Literal(objType.strip().replace("_", " "))))
                except AttributeError:
                    print(objType)

                # Create catalogue item IRI
                catItemIRI = slb[f'E73/bm_dataset/{recordID}/{catItemCount}'];
                catItemCount += 1
                graph.add((catItemIRI, RDFS.label, Literal(f'BM catalogue item {recordID}/{catItemCount}')))

                # Connect catalogue item to record
                graph.add((recordIRI, ecrm.P106_is_composed_of, catItemIRI))

                # Connect inscription to entry
                graph.add((catItemIRI, slb.SL10_specifically_refers_to, typeIRI))

        # For each title...
        for objTitle in str(row["Title"]).split('~'):
            
            if objTitle == 'nan':
                continue

            if objTitle.strip().replace(" ", "_"):
                # Create title IRI
                try:
                    objTitle = objTitle.translate(str.maketrans({'"': None}));
                    titleIRI = slb[f'E35/{objTitle.strip().replace(" ", "_").replace("[", "").replace("|", "").replace("]", "")}']
                    if not rfc3987.parse(titleIRI)['IRI']:
                        print(titleIRI)
                    graph.add((titleIRI, RDF.type, ecrm.E35_Title))
                    graph.add((titleIRI, RDFS.label, Literal(objTitle.strip().replace("_", " "))))
                except AttributeError:
                    print(objTitle)

                # Create catalogue item IRI
                catItemIRI = slb[f'E73/bm_dataset/{recordID}/{catItemCount}'];
                catItemCount += 1
                graph.add((catItemIRI, RDFS.label, Literal(f'BM catalogue item {recordID}/{catItemCount}')))

                # Connect catalogue item to record
                graph.add((recordIRI, ecrm.P106_is_composed_of, catItemIRI))

                # Connect inscription to entry
                graph.add((catItemIRI, slb.SL10_specifically_refers_to, titleIRI))
    graph.serialize(destination=f'/Users/danielemetilli/Downloads/bm-data-pt-{int(j/BATCH)+1}.ttl', format="turtle");
    print(int(j/BATCH)+1)
    j += BATCH
```
